# Logistic Map

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def logmap(r,x):
    p=[]
    for i in range(0,40):
        xi=r*x*(1-x)
        x=xi
        p.append(xi)
    y=[i for i in range(0,40)]
    #print(y)
    #print(p)
    plt.plot(y,p)
    plt.xlabel('Iteration')
    plt.ylabel('X')
```

```python
logmap(3.52,0.2)
```

```python
logmap(0.9,0.2)
```

```python
logmap(4,0.2)
```

```python
logmap(2.6,0.2)
```

# Tent Map

```python
def tentmap(a,b):
    c=[]
    for i in range(1,40):
        if 0<=a<=0.5:
            ai=2*b*a
            a=ai
        elif 0.5<a<=1:
            ai=2*b*(1-a)
            a=ai
        c.append(ai)
    d=[i for i in range(1,40)]
    #print(d)
    #print(c)
    plt.plot(d,c)
    plt.xlabel('Iteration')
    plt.ylabel('X') 
```

```python
tentmap(0.5,0.45)
```

```python
tentmap(0.5,0.2)
```

```python
tentmap(0.5,0.5)
```

```python
tentmap(0.5,0.75)
```
